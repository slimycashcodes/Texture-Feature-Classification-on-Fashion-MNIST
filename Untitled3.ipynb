```python
import numpy as np
import time
from tensorflow.keras.datasets import fashion_mnist
from skimage.feature import hog
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report

print("loading fashion mnist..")
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

x_train_n = x_train.astype('float32')/255.0
x_test_n = x_test.astype('float32') / 255.0

def extractHog(imgs):
    feats = []
    for img in imgs:
        fd = hog(img, orientations= 9, pixels_per_cell = (8,8), cells_per_block = (2,2), visualize=False)
        feats.append(fd)

    return np.array(feats)

print("extracting hog features from train... ")
train_features = extractHog(x_train_n)

print("extracting test features now ")
test_features = extractHog(x_test_n)

print("\ntraining naive bayes")
nb = GaussianNB()
st = time.time()
nb.fit(train_features, y_train)
nb_time = time.time() - st

nb_preds = nb.predict(test_features)
nb_acc = accuracy_score(y_test, nb_preds)

print("naive bayes accuracy: ", round(nb_acc * 100, 2), "%")
print("time: ", round(nb_time, 2), "s")
print(classification_report(y_test, nb_preds))


print("\ntraining random forest...")
rf = RandomForestClassifier(n_estimators=100, random_state=42)
st2 = time.time()
rf.fit(train_features, y_train)
rf_time = time.time() - st2

rf_preds = rf.predict(test_features)
rf_acc = accuracy_score(y_test, rf_preds)

print("random forest accuracy: ", round(rf_acc * 100, 2), "%")
print("time: ", round(rf_time, 2), "s")
print(classification_report(y_test, rf_preds))
```
